```python
class LinearCongruentialGenerator:
    def __init__(self, seed, a, c, m):
        """
        Initialize the Linear Congruential Generator with the given parameters.

        :param seed: The initial value (seed)
        :param a: Multiplier (a)
        :param c: Increment (c)
        :param m: Modulus (m)
        """
        self.seed = seed
        self.a = a
        self.c = c
        self.m = m
        self.state = seed  # initial state

    def generate(self):
        """
        Generate the next random number in the sequence.

        :return: The next random number in the sequence
        """
        self.state = (self.a * self.state + self.c) % self.m
        return self.state / self.m  # Normalize to [0, 1)

# Parameters for the LCG
seed = 42      # Initial state
a = 1664525    # Multiplier
c = 1013904223 # Increment
m = 2**32      # Modulus (a large prime number or power of 2)

# Create an instance of the LinearCongruentialGenerator
lcg = LinearCongruentialGenerator(seed, a, c, m)

# Generate 10 random numbers
for _ in range(10):
    print(lcg.generate())
```

```python
import numpy as np
def lcg(seed, a, c, m, size):
    """
    Linear Congruential Generator (LCG) function.

    Parameters
    ----------
    seed : `int`
        The initial value (X0) for the LCG.
    a : `int`
        The multiplier.
    c : `int`
        The increment.
    m : `int`
        The modulus.
    size : `int`
        The number of random numbers to generate.

    Returns
    -------
    `np.ndarray[float]`
        A NumPy array containing the generated sequence.
    """
    # Initialize the sequence array with zeros
    sequence = np.zeros(size)

    # Set the initial value
    sequence[0] = seed

    # Generate the sequence
    for i in range(1, size):
        sequence[i] = (a * sequence[i - 1] + c) % m

    return sequence
```

```python
import numpy as np
import matplotlib.pyplot as plt

def lcg(seed, a, c, m, size):
    """
    Linear Congruential Generator (LCG) function.

    Parameters
    ----------
    seed : `int`
        The initial value (X0) for the LCG.
    a : `int`
        The multiplier.
    c : `int`
        The increment.
    m : `int`
        The modulus.
    size : `int`
        The number of random numbers to generate.

    Returns
    -------
    `np.ndarray[float]`
        A NumPy array containing the generated sequence.
    """
    # Initialize the sequence array with zeros
    sequence = np.zeros(size)

    # Set the initial value
    sequence[0] = seed

    # Generate the sequence
    for i in range(1, size):
        sequence[i] = (a * sequence[i - 1] + c) % m

    # Normalize to [0, 1)
    return sequence / m

# Parameters for the LCG
seed = 42      # Initial state
a = 1664525    # Multiplier
c = 1013904223 # Increment
m = 2**32      # Modulus (a large prime number or power of 2)
size = 10000   # Number of random numbers to generate

# Generate the sequence
random_numbers = lcg(seed, a, c, m, size)

# Plotting the histogram to visualize the distribution
plt.figure(figsize=(8, 6))
plt.hist(random_numbers, bins=50, density=True, alpha=0.7, color='blue')
plt.title("Histogram of Linear Congruential Generator Output")
plt.xlabel("Generated Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
